```python
def get_qus(text):
    test_fea = model.transform(text)
    sim_list = np.empty([189])
    for i in range(189):
        sim_list[i] = cosine_similarity(test_fea,features[i])
    x = sim_list.argsort()
    return x[-5:]
```
